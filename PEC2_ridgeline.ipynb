```python
# Se importa la librería de pandas
import pandas as pd
# Carga los datos desde un archivo Excel que se han extraído del INE
archivo_excel = r"C:\Users\ruben\Desktop\UOC\ASIGNATURAS OBLIGATORIAS\M2.859 - Visualización de datos\PEC2\31933.xlsx"
# Reemplaza con la ruta de tu archivo
datos = pd.read_excel(archivo_excel)
# Muestra las primeras filas para verificar la carga
print(datos.head())
```

```python
# Crea el dataframe
df = pd.DataFrame(datos)
```

```python
# Se muestran los datos cargados en el dataframe
df
```

```python
# Crea el gráfico de joyplot
fig, axes = joypy.joyplot(df, by="Periodo", column="Total")

# Se itera sobre la lista de ejes y se configura cada uno
for i, ax in enumerate(axes):
    ax.spines['left'].set_visible(True)  # Mostrar eje izquierdo (y)
    ax.spines['bottom'].set_visible(True)  # Mostrar eje inferior (x)
    ax.set_xlabel("Número de nacimientos")  # Etiqueta eje X

    # Solo se agrega la etiqueta del eje Y en el cuarto gráfico
    if i == 3:
        ax.set_ylabel("Año")  # Etiqueta eje Y

# Muestra el gráfico de joyplot
plt.show()
```
